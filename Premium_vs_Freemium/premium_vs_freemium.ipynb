# Premium vs Freemium

Find the total number of downloads for paying and non-paying users by date.<br>
Include only records where non-paying customers have more downloads than paying customers. <br>
The output should be sorted by earliest date first and contain 3 columns date, non-paying downloads, paying downloads.

```python
# Import your libraries
import pandas as pd
```

```python
# ms_user_dimension = pd.read_clipboard()
# ms_user_dimension.to_csv('ms_user_dimension.csv',index=False)
ms_user_dimension = pd.read_csv('ms_user_dimension.csv')
ms_user_dimension.head()
```

```python
# ms_acc_dimension = pd.read_clipboard()
# ms_acc_dimension.to_csv('ms_acc_dimension.csv',index=False)
ms_acc_dimension = pd.read_csv('ms_acc_dimension.csv')
ms_acc_dimension.head()
```

```python
# ms_download_facts = pd.read_clipboard()
# ms_download_facts.to_csv('ms_download_facts.csv',index=False)
ms_download_facts = pd.read_csv('ms_download_facts.csv')
ms_download_facts.head()
```

```python
# First Merge 
merge1 = ms_user_dimension.merge(ms_acc_dimension, on='acc_id')
merge1.head()
```

```python
# Second Merge
merge2 = merge1.merge(ms_download_facts, on='user_id')
merge2.head()
```

```python
# Create pivot table
pivoted = merge2.pivot_table(index='date', columns='paying_customer', values='downloads', aggfunc='sum')
pivoted.head()
```

```python
# Keep only records where theres more downloads for non paying customers
final = pivoted[pivoted['no'] > pivoted['yes']].reset_index()
final
```

```python
# Remove named index
final = final.rename_axis(None, axis=1)
final
```
