Find the details of each customer regardless of whether the customer made an order. Output the customer's first name, last name, and the city along with the order details.
You may have duplicate rows in your results due to a customer ordering several of the same items. Sort records based on the customer's first name and the order details in ascending order.

```python
# Import your libraries
import pandas as pd
```

```python
# customers = pd.read_clipboard()
# customers.to_csv('customers.csv',index=False)
customers = pd.read_csv('customers.csv')
customers.head()
```

```python
customers.isnull().values.any()
```

```python
# orders = pd.read_clipboard()
# orders.to_csv('orders.csv',index=False)
orders = pd.read_csv('orders.csv')
orders.head()
```

```python
orders.isnull().values.any()
```

```python
# Grab only the two columns we need from the orders df
orders = orders[['cust_id','order_details']]
orders.head()
```

```python
# Rename cust-id to id for merge
orders.rename(columns = {'cust_id':'id'}, inplace = True)
orders.head()
```

```python
# Merge customers with orders left on id to include missing right values
merged = customers.merge(orders, on='id', how='left')
merged.head()
```

```python
# Reduce dataframe to columns requested
merged = merged[['id','first_name','last_name','city','order_details']]
merged.head()
```

```python
# Sort the values by first name
sorted = merged.sort_values(by=['first_name'])
sorted
```
