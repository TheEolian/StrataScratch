# Highest Salary In Department

Find the employee with the highest salary per department.
Output the department name, employee's first name along with the corresponding salary.

```python
# Import your libraries
import pandas as pd
```

```python
#employee = pd.read_clipboard()
#employee.to_csv('employee.csv',index=False)
employee = pd.read_csv('employee.csv')
employee.head()
```

```python
# Check how many departments are in the dataset
employee.department.value_counts()
```

### Three departments needed: Sales, Audit, and Management

```python
# create datefrace of sales department, sory if by highest salary first and only keep the top row
sales = employee.loc[employee['department'] == "Sales"].sort_values(by=['salary'], ascending=False)
sales = sales.head(1)
sales
```

```python
# create datefrace of management department, sory if by highest salary first and only keep the top row
management = employee.loc[employee['department'] == "Management"].sort_values(by=['salary'], ascending=False)
management = management.head(1)
management
```

```python
# create datefrace of audit department, sory if by highest salary first and only keep the top row
audit = employee.loc[employee['department'] == "Audit"].sort_values(by=['salary'], ascending=False)
audit = audit.head(1)
audit
```

```python
# Concat the 3 dataframes of each departments top salary together
final = pd.concat([sales, management, audit], axis=0)
```

```python
final
```

```python
# Return only the department name, first name of employee, and salary
final[['department','first_name','salary']]
```
