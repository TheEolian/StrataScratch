### Classify Business Type

Classify each business as either a restaurant, cafe, school, or other. A restaurant should have the word 'restaurant' in the business name. For cafes, either 'cafe', 'caf√©', or 'coffee' can be in the business name. 'School' should be in the business name for schools. All other businesses should be classified as 'other'. Output the business name and the calculated classification.

```python
# Import your libraries
import pandas as pd
```

```python
# set display options
pd.options.display.max_columns = 20
pd.options.display.max_rows = 20
pd.options.display.max_colwidth = 80
```

```python
#sf_restaurant_health_violations = pd.read_clipboard()
#sf_restaurant_health_violations.to_csv('sf_restaurant_health_violations.csv',index=False)
sf_restaurant_health_violations = pd.read_csv('sf_restaurant_health_violations.csv')
sf_restaurant_health_violations.head()
```

```python
# Create new business type column with default 'z'
sf_restaurant_health_violations['business_type'] = 'z'
```

```python
# create df of only restaurants
restaurants = sf_restaurant_health_violations.loc[sf_restaurant_health_violations['business_name'].str.contains("restaurant", na=False, case=False)]
restaurants.head()
```

```python
# create df of only cafes
cafe = sf_restaurant_health_violations.loc[sf_restaurant_health_violations['business_name'].str.contains("cafe", na=False, case=False)]
caf√© = sf_restaurant_health_violations.loc[sf_restaurant_health_violations['business_name'].str.contains("caf√©", na=False, case=False)]
coffee = sf_restaurant_health_violations.loc[sf_restaurant_health_violations['business_name'].str.contains("coffee", na=False, case=False)]
cafes = pd.concat([cafe, caf√©, coffee], axis=0)
```

```python
cafes.head()
```

```python
# Create dataframe of only schools
school = sf_restaurant_health_violations.loc[sf_restaurant_health_violations['business_name'].str.contains("school", na=False, case=False)]
```

```python
# create business type columns for each df
restaurants['business_type'] = 'restaurant'
cafes['business_type'] = 'cafe'
school['business_type'] = 'school'
```

```python
# concat all dataframes
final = pd.concat([sf_restaurant_health_violations, restaurants, cafes,school], axis=0)
```

```python
# sort by business name the business type
final.sort_values(by=['business_name', 'business_type'], ascending=[True, True],inplace=True)
```

```python
# drop duplicates on business name, this will drop any of the 'z' business tpyes if they are present witha  different value
final = final.drop_duplicates(subset=['business_name'])
```

```python
# Rename business tpye 'z' to 'other'
final['business_type'] = final['business_type'].replace({'z': 'Other'})
```
